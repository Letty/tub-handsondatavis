### Setup von hilfreichen Python Packages

```r
library(ggplot2)   ## graphics plotting package
```

### Import der Daten

```r
my_data <- read.csv("owid-covid-data.csv", header=TRUE, sep = "," )
my_data['date'] <- as.Date(my_data$date) # Umwandeln von string format zu datetime format

# Reduktion der Datenmenge - beispiel daten für GDP aller Länder in Europa 
data_sample <- my_data[!duplicated(my_data$location), ] 
data_gdp = data_sample[,c('iso_code', 'continent', 'location', 'population_density', 'median_age', 'aged_65_older',
       'aged_70_older', 'gdp_per_capita', 'extreme_poverty', 'life_expectancy')]
data_gdp <- data_gdp[data_gdp$continent == "Europe", ] 

data_austria <- my_data[my_data$location == "Austria", ] 
```

# Bar chart

```r
# wie viele einträge haben wir pro kontinent? als Graphik
ggplot(data = my_data) + geom_bar(mapping = aes(x = continent))
```

# Histogram

```r
ggplot(data_gdp, aes(x=gdp_per_capita)) + geom_histogram()
```

# Boxplot

```r
ggplot(my_data, aes(x=continent, y=gdp_per_capita)) + geom_boxplot(mapping = aes(x = continent))
```

# Scatterplot

```r
ggplot(data_gdp, aes(x=gdp_per_capita, y=life_expectancy)) + geom_point()
```

# Line chart

```r
ggplot(data_austria, aes(x=date, y=new_tests)) + geom_line()
```

# Titel und Achsenbeschriftung

```r
ggplot(data_gdp, aes(x=gdp_per_capita, y=life_expectancy)) + geom_point() + ggtitle("Vergleich GDP und Lebenserwartung in Europa")
```

# Farbänderungen

```r
ggplot(data_gdp, aes(x=gdp_per_capita, y=life_expectancy, color=location)) + geom_point(size=6)
```

```r
ggplot(my_data, aes(x=continent, y=gdp_per_capita, color=continent)) + geom_boxplot(mapping = aes(x = continent))
```

# Export

```r
ggplot(data_gdp, aes(x=gdp_per_capita, y=life_expectancy)) + geom_point() + ggtitle("Vergleich GDP und Lebenserwartung in Europa")
ggsave("R - scatterplot example.svg")
```

```r
ggplot(my_data, aes(x=continent, y=gdp_per_capita, color=continent)) + geom_boxplot(mapping = aes(x = continent))
ggsave("R - Boxplot example.svg")
```
